# 物元法

（资料很少，部分概念为个人理解，应该没问题）（概念统一解释）

**核心思想**：

物元法对多指标的对象使用【综合关联度】对人为划分的【等级】做出关联程度的量化评价，是一种对【不相容问题】的解决方法

**步骤**：

1. 确定【物元】
2. 确定【经典域】
3. 确定【节域】
4. 确定【待评物元】
5. 确定【关联函数】
6. 计算【关联度】
7. 确定【权重】
8. 计算【综合关联度】


**概念解释**：

1. 【不相容问题】：当提供的条件不能达到要实现的目标时，称为不相容问题（曹冲称象）

   在这里可以理解为：我们已有的数据不足以支持我们对某一对象做出评价

2. 【物元】：把受评价的对象用“事物、特征、量值” 3 个要素以有序三元组的形式来描述，这个三元组就叫物元。

   将要评价的事物记作 $N$，其特征（指标）记为 $c$，特征值记为 $v$，

   假设 $N$ 有 $n$ 个特征: $c_1, c_2, \ldots, c_n$，则这 $n$ 个特征对应的量值为:$v_1, v_2, \ldots, v_n$

   而物元则表示为：

   $$
   R=\left[\begin{array}{ccc}
   N & c_1 & v_1 \\
   & c_2 & v_2 \\
   & \vdots & \vdots \\
   & c_n & v_n
   \end{array}\right]=\left[\begin{array}{c}
   R_1 \\
   R_2 \\
   \vdots \\
   R_n
   \end{array}\right]
   $$

   简记为 $\boldsymbol R=(N, \boldsymbol c, \boldsymbol v)$

   其中 $\boldsymbol c=(c_1, c_2, \ldots, c_n)$，$\boldsymbol v=(v_1, v_2, \ldots, v_n)$

3. 【节域】：根据事物的特征值所在的总区间确定

   对某一事物 $N_p$ 而言，其每个特征 $c_i$ 都应该有一个总的取值范围 $v_{p i}$，用区间 $(a_{p,i},\,b_{p,i})$表示。

   三者组合称为【节域物元】$R_p$：

   $$
   R_p=\left[\begin{array}{ccc}
   N_p & c_1 & v_{p 1} \\
   & c_2 & v_{p 2} \\
   & \vdots & \vdots \\
   & c_n & v_{p n}
   \end{array}\right]=\left[\begin{array}{ccc}
   N_p & c_1 & \left(a_{p 1}, b_{p 1}\right) \\
   & c_2 & \left(a_{p 2}, b_{p 2}\right) \\
   & \vdots & \vdots \\
   & c_n & \left(a_{p n}, b_{p n}\right)
   \end{array}\right]
   $$

4. 【经典域】：根据【等级】下事物的特征值所在的区间确定

   我们人为地将评价等级划分为 $m$ 级，因此事物可用某一等级 $j$ 下的待评对象表示，即 $N_j$，

   那么事物的每个特征 $c_i$ 都应该有一个取值范围 $v_{j i}$，用区间 $(a_{j,i},\,b_{j,i})$表示。

   这三者组合起来就称为【经典域物元】$R_j$：

   $$
   R_j=\left[\begin{array}{ccc}
   N_j & c_1 & v_{j 1} \\
   & c_2 & v_{j 2} \\
   & \vdots & \vdots \\
   & c_n & v_{j n}
   \end{array}\right]=\left[\begin{array}{ccc}
   N_j & c_1 & \left(a_{j 1}, b_{j 1}\right) \\
   & c_2 & \left(a_{j 2}, b_{j 2}\right) \\
   & \vdots & \vdots \\
   & c_n & \left(a_{j n}, b_{j n}\right)
   \end{array}\right]
   $$

   因此，节域对应的区间相当于每个等级下的经典域对应的区间的并集。

5. 【待评物元】：

   设有 $y$ 个待评事物，则待评对象 $N_x$ 的物元表示为 $R_x$：

   $$
   R_x=\left[\begin{array}{ccc}
   N_x & c_1 & v_1 \\
   & c_2 & v_2 \\
   & \vdots & \vdots \\
   & c_n & v_n
   \end{array}\right] \quad x=1 \cdots y
   $$

6. 【关联函数】：

   将有界区间 $V=[a,\,b]$ 的模定义为：$|V|=|b-a|$

   某一点 $v_{x}$ 到区间 ${V}$ 的距离为:

   $$
   \rho(v_{x}, V)=\left|v_{x}-\frac{1}{2}(a+b)\right|-\frac{1}{2}(b-a)
   $$

   则关联函数定义为：

   $$
   k(v_x)= \left\{\begin{aligned}
   & \frac{-\rho(v_x, V)}{|V|} & (v_x \in V) \\
   & \frac{\rho(v_x, V)}{\rho(v_x, V_{总})-\rho(v_x, V)} & (v_x \notin V)
   \end{aligned} \right.
   $$

7. 【关联度】：关联函数的表达式为：

   $$
   k_{x j}(v_i)= \left\{\begin{aligned}
   & \frac{-\rho(v_{x i}, v_{j i})}{|v_{j i}|} & (v_{x i} \in v_{j i}) \\
   & \frac{\rho(v_{x i}, v_{j i})}{\rho(v_{x i}, v_{p i})-\rho(v_{x i}, v_{j i})} & (v_{x i} \notin v_{j i})
   \end{aligned} \right.
   $$

   其中，$k_{x j}(v_i)$ 表示第 $x$ 个待评物元，第 $j$ 个等级，第 $i$ 个指标的【关联度】

   理解为：待评事物 $N_x$ 的第 $i$ 个指标与第 $j$ 个等级的关联程度

   因为根据理解，一共有 y 个待评事物，每个事物都有 n 个指标，又因为要和 $m$ 个等级计算，所以最终应当计算得到 $y × n × m$ 个关联度，制表如下：

8. 【权重】：权重的获得方法各异，使用熵权法，该模型就称为熵权可拓物元模型。
9. 【综合关联度】：评价结果，与某个等级关联程度越高，视为其水平越可能在该等级
   $$K_j\left(N_x\right)=\sum_{i=1}^n \omega_i k_j\left(v_i\right)$$
   理解为：待评事物 $N_x$ 与第 $j$ 个等级的关联程度，用权重将指标统合起来。

   最终应该得到 $y × m$ 个综合关联度，制表如下：

假设指标：人均综合月收入，经典域分别为：水平高(30000, 40000), 较高(20000, 30000), 中等(9000, 20000), 很低(5000, 9000)

高等学校在校学生数，经典域分别为：水平高(45, 60), 较高(30, 45), 中等(15, 30), 很低(1, 15)

人均综合月收入的节域就是 $v_{p,\,income}=(5000, 40000)$

高等学校在校学生数的节域就是 $v_{p,\,student}=(1, 60)$


```python
import numpy as np
```

```python
# 对象 y - x
obj = ["A", "B", "C", "D", "E"]
y = len(obj)
# 等级 m - j
level = ["高", "较高", "中等", "很低"]
m = len(level)
# 指标 n - i
c = ["income", "student"]
n = len(c)

# 指标量值 vx
vx_income = [35549, 24981, 24420, 17760, 8490]
vx_student = [59.30, 52.71, 43.64, 18.05, 8.53]
vx = [vx_income, vx_student]

# 经典域 vj
vj_income = [(30000, 40000), (20000, 30000), (9000, 20000), (5000, 9000)]
vj_student = [(45, 60), (30, 45), (15, 30), (1, 15)]
vj = [vj_income, vj_student]

# 节域 vp
vp_income = (5000, 40000)
vp_student = (1, 60)
vp = [vp_income, vp_student]
```

```python
def is_in(v, V):
    # The borderline cases are the same, so it doesn't matter
    if v > V[0] and v < V[1]:
        return True
    else:
        return False


def rho(v, V):
    return abs(v - 0.5 * (V[0] + V[1])) - 0.5 * (V[1] - V[0])


def k(vx, vj, vp, x, i, j):
    # variable quantities
    vxi = vx[i][x]
    vji = vj[i][j]
    vpi = vp[i]
    rho_xj = rho(vxi, vji)
    rho_xp = rho(vxi, vpi)
    # if vxi is in vji
    flag = is_in(vxi, vji)
    if flag:
        return -rho_xj/(vji[1]-vji[0])
    elif not flag:
        return rho_xj/(rho_xp-rho_xj)
```

```python
print("\t\t", end='')
for x in range(y):
    print("{}\t\t\t\t".format(obj[x]), end='')
print("\n\t\t", end='')
for x in range(y):
    for j in range(m):
        print("{}\t".format(level[j]), end='')
print()

for i in range(n):
    print("{}\t\t".format(c[i]), end = '')
    for x in range(y):
        for j in range(m):
            print("{}\t".format(round(k(vx, vj, vp, x, i, j), 2)), end='')
    print()
# for i in range(n):
#     print("\t\t", end = '')
#     for x in range(y):
#         print("{}\t\t\t\t".format(vx[i][x]), end='')
#     print()
# for i in range(n):
#     print("\t\t", end = '')
#     for j in range(m):
#         print("{}\t".format(vj[i][j]), end='')
#     print()
# print()
```
