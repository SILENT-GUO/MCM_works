```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_fungi_temperature_original = pd.read_csv('./Fungi_temperature_curves.csv')
namelist = ["p.sang.s", "t.chion.n", "x.sub.s"]
df_fungi_temperature_needed = pd.DataFrame()
species, temp_c, hyphal_rate = [], [], []
for i in range(len(df_fungi_temperature_original["species"])):
    name = df_fungi_temperature_original["species"][i]
    if name in namelist:
        species.append(name)
        temp_c.append(df_fungi_temperature_original["temp_c"][i])
        hyphal_rate.append(df_fungi_temperature_original["hyphal_rate"][i])
df_fungi_temperature_needed["species"] = species
df_fungi_temperature_needed["temp_c"] = temp_c
df_fungi_temperature_needed["hyphal_rate"] = hyphal_rate
    
grouped = df_fungi_temperature_needed.groupby('species')
df_fungi_temperature = pd.DataFrame()
for name, group in grouped:
    group = group.reset_index(drop=True)
    df_fungi_temperature["temp_c_" + name] = group["temp_c"]
    df_fungi_temperature["hyphal_rate_" + name] = group["hyphal_rate"]
for name in namelist:
    plt.scatter(df_fungi_temperature['temp_c_' + name],
                df_fungi_temperature['hyphal_rate_' + name], s=1, label=name)
plt.legend()
plt.show()
```
